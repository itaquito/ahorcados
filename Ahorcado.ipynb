```python
# Manda el menú para seleccionar las palabras, también escucha y valida las respuestas.
# Regresa la palabra con la que se va a jugar. 
def seleccionarPalabra():
    print(agregarBarras(
        "¡Vale! ¿Con qué palabras quieres jugar? Escribe el número de una opción para seleccionarla.\n" +
        "Nota: Las mayúsculas, mínusculas, vocales con y sin acento no se distinguen en las palabras.\n" +
        "\n" +
        "1) Palabras predeterminadas: Juega con las palabras predeterminadas, podrás seleccionar una categoría en el siguiente menú.\n" +
        "2) Palabras personalizadas: Ingresa tus propias palabras o frases."
    ))
    respuesta = input("Selecciona una opción (1-2): ")
    
    if (respuesta == "1"): # Opción 1 (Palabras predeterminadas)
        # Aplicamos title() a cada una de las categorías para que se vean mejor.
        listaCategorias = ", ".join(list(DIC_PALABRAS))
            
        print(agregarBarras(
            "Escribe el nombre de una categoría de palabras para seleccionarla.\n" +
            "Categorías disponibles: " + listaCategorias
        ))
        
        # Seguir preguntando hasta que recibamos una respuesta válida
        while True:
            respuesta = input("Escribe el nombre de una categoría: ")
            respuestaEstandarizada = quitarAcentos(respuesta) # Eliminar acentos y mayúsculas
            if respuestaEstandarizada not in DIC_PALABRAS:
                print(agregarBarras(
                    "¡Respuesta inválida! Vuelve a intentarlo\n" +
                    "Categorías disponibles: " + listaCategorias + "\n\n" +
                    "Tip: Escribe el nombre de la categoría exactamente como aparece arriba, incluyendo espacios."
                ))
                continue
            
            palabrasSeleccionadas = DIC_PALABRAS[respuestaEstandarizada]
            break
                
    elif (respuesta == "2"): # Opción 2 (Palabras personalizadas)
        # Seguir preguntando hasta que recibamos una respuesta válida
        while True:
            print(agregarBarras("Escribe todas las palabras o frases que quieras separadas por una diagonal (/). Seleccionaremos una al azar para el juego."))
            respuesta = input("Escribe las palabras o frases aquí: ")
            palabrasSeleccionadas = respuesta.split("/")
            
            # Asegurarnos de que haya al menos una letra
            if len(respuesta) < 1:
                print("\n¡Tienes que inserta al menos una palabra! Vuelve a intentarlo...")
                continue
            break    
    else:
        print(agregarBarras("¡Esa opción no existe! Vuelve a intentarlo..."))
        return seleccionarPalabra()
    
    return random.choice(palabrasSeleccionadas)
```

```python
# Manda el menú para seleccionar el número de jugadores y sus nombres, también escucha y valida las respuestas.
# Regresa el nombre de los jugadores en una lista. 
def seleccionarNombresJugadores():
    respuesta = input(agregarBarras("¿Cuántos jugadores van a jugar?"))
    
    # Checar que sea un número entero
    if not respuesta.isdigit():
        print("\n¡Ese no es un número entero! Vuelve a intentarlo...")
        return seleccionarNombresJugadores()
    
    numeroJugadores = int(respuesta)
    
    if (numeroJugadores < 2) or (numeroJugadores > 100):
        print("\nSe necesitan mínimo 2 jugadores y máximo 10. Vuelve a intentarlo...")
        return seleccionarNombresJugadores()
    
    jugadores = []
    for x in range(numeroJugadores):
        # Seguir preguntando hasta que tengamos una respuesta válida
        while True:
            respuesta = input("¿Cómo se llama el jugador #" + str(x + 1) + "? ")
            
            # Checar que sea de la longitud correcta
            if len(respuesta) < 1 or len(respuesta) > 10:
                print(agregarBarras("Tu nombre sólo puede tener de 1 a 10 caracteres de largo."))
                continue
            
            # Checar que no contenga "/"
            if "/" in respuesta:
                print(agregarBarras("Tu nombre tiene caracteres inválidos."))
                continue
            
            # Checar que no sea un nombre duplicado
            if respuesta in jugadores:
                print(agregarBarras("Tu nombre ya está en uso."))
                continue
            
            jugadores.append(respuesta)
            break
            
    return jugadores
        
```

```python
# Genera un texto con la palabra escondida, mostrando las letras adivinadas y los caracteres prohibidos. Ejemplo: H_la, b_enas ta_des.
# Toma como argumento la palabra a adivinar, y las letras adivinadas (en lista, en minúsculas y sin acentos).
# Regresa la palabra escondida y un valor booleano indicando si ya se adivinó toda la palabra
def generarPalabraEscondida(palabraAdivinar, letrasAdivinadas):
    palabraEscondida = ""
    palabraAdivinada = True

    for letra in palabraAdivinar:
        letraEstandarizada = quitarAcentos(letra) # Quitar acentos y mayúsculas.
        
        if (letraEstandarizada in letrasAdivinadas) or (letraEstandarizada not in CARACTERES_PERMITIDOS): # Si la letra ya se adinivó o NO es un caracter permitido, mostrarla.
            palabraEscondida += letra
        else:
            palabraEscondida += "_"
            palabraAdivinada = False # Si ponemos un "_" significa que al menos falta una letra por adivinar
    
    return({
        "palabraAdivinada": palabraAdivinada,
        "palabraEscondida": palabraEscondida
    })
```

```python
# Una función muy sencilla para regresar el texto de entrada con dos barras. Una arriba y otra abajo.
# Ayuda a diferenciar las secciones de texto que manda el programa
def agregarBarras(texto):
    return(
        "\n━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n\n" +
        texto +
        "\n\n━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n"
    )
```

```python
# Quita acentos y mayúsculas a un string.
def quitarAcentos(texto):
    textoLower = texto.lower()
    textoFinal = ""
    for caracter in textoLower:
        if (caracter == "á"):
            textoFinal += "a"
        elif (caracter == "é"):
            textoFinal += "e"
        elif (caracter == "í"):
            textoFinal += "i"
        elif (caracter == "ó"):
            textoFinal += "o"
        elif (caracter == "ú"):
            textoFinal += "u"
        else:
            textoFinal += caracter # No tiene acento, pasa la letra normal.
    return textoFinal
```

```python
# Pregunta el nombre y guarda el tiempo en el archivo de ranking. 
# Toma como argumentos la palabra adivinada, el tiempo en segundos y las fallas.
# El argumento "nombre" es opcional. Si no se establece, la función preguntará por el nombre y mandará el mensaje de confirmación.
def guardarTiempo(palabraAdivinada, tiempo, fallas, nombre=False):
    # No tenemos nombre
    if not nombre:
        print(
            "\nEscribe tu nombre para guardar tu tiempo.\n" +
            "O presiona ENTER sin escribir nada para no guardar tu tiempo.\n"
        )
        respuesta = input("Tu nombre: ")

        # No guardar el tiempo
        if len(respuesta) == 0:
            print("No se guardó tu tiempo.")
            return

        # Checar que sea de la longitud correcta
        if len(respuesta) > 10:
            print(agregarBarras("Tu nombre sólo puede tener 10 caracteres de largo."))
            return guardarTiempo(palabraAdivinada, tiempo, fallas)

        # Checar que no contenga "/"
        if "/" in respuesta:
            print(agregarBarras("Tu nombre tiene caracteres inválidos."))
            return guardarTiempo(palabraAdivinada, tiempo, fallas)

    # Abrimos el archivo en append, también crea el archivo si no existe.
    with open("JuegoNormal.ranking", "a") as archivo:
        archivo.write(respuesta if not nombre else nombre + "/" + palabraAdivinada + "/" + str(tiempo) + "/" + str(fallas) + "\n")
        if not nombre:
            print("Se guardó tu tiempo correctamente.")
    return
```

```python
# Obtiene el estado del juego multijugador
# Nos sirve para saber si ya hay un jugador que ganó o si todos los jugadores perdieron.
# jugadorGanador > Es True si al menos un jugador ya adivinó la palabra
# jugadoresPerdedores > Es True si todo los jugadores perdieron
# Si al menos uno de los valores de arriba es True, se deberá acabar el juego.
# Regresa True si el juego puede seguir y False si se debe acabar.
def obtenerEstadoMultijugador(infoJugadores):
    jugadorGanador = False
    jugadoresPerdedores = False
    
    cantidadJugadores = len(infoJugadores)
    for jugador in infoJugadores:
        if jugador["palabraAdivinada"]:
            jugadorGanador = True
        elif jugador["fallas"] >= (len(HOMBRE_PALO) - 1):
            cantidadJugadores -= 1
    # Ya no quedan jugadores sin perder
    if cantidadJugadores <= 0:
        jugadoresPerdedores = True
    
    return(not (jugadorGanador or jugadoresPerdedores))
```

```python
# Genera el texto de los ganadores, también guarda el tiempo de los ganadores en el ranking
# Toma como argumento la lista de jugadores y la palabra usada en ese juego.
# Regresa un texto mostrando los ganadores.
def generarGanadores(infoJugadores, palabraSeleccionada):
    # Dos listas para guardar los jugadores que ganaron y los que no
    ganadores = []
    noGanadores = []
    for jugador in infoJugadores:
        if jugador["palabraAdivinada"]:
            ganadores.append(
                jugador["nombre"] +
                " en " + str(jugador["tiempo"]) + " segundos"
            )
            guardarTiempo(palabraSeleccionada, jugador["tiempo"], jugador["fallas"], jugador["nombre"])
        else:
            noGanadores.append(
                jugador["nombre"] +
                " > " + generarPalabraEscondida(palabraSeleccionada, jugador["letrasAdivinadas"])["palabraEscondida"] +
                " > " + str(jugador["tiempo"]) + " segundos"
            )
    
    usarSintetizador("¡Felicidades! La palabra o frase correcta era " + palabraSeleccionada + ". Y los ganadores son: " + ", ".join(ganadores))
    return(agregarBarras(
        "¡Se acabó el juego! La palabra o frase correcta era " + palabraSeleccionada + "\n"
        "Ganador" + ("es" if len(ganadores) != 1 else "") +
        "\n" +
        ", ".join(ganadores) +
        "\n\n"
        "Demás jugadores:\n" +
        "\n".join(noGanadores)
    ))
```

```python
# Genera el ranking desde el archivo "JuegoNormal.ranking"
# Regresa el texto con el ranking
def generarRanking():
    #Crear el archivo del ranking si no existe
    archivo = open("JuegoNormal.ranking","a")
    archivo.close()

    # Abrimos el archivo con Read.
    with open("JuegoNormal.ranking", "r") as archivo: 
        lineas = archivo.readlines()
        
        # No hay líneas
        if len(lineas) == 0:
            return("¡No hay ningúna partida guardada! Juega un poco para guardar un tiempo.")
        
        # Convertir las lineas a un formato que podamos usar más fácilmente (una lista de diccionarios)
        ranking = []
        for linea in lineas:
            argumentos = linea.split("/")
            ranking.append({
                "nombre": argumentos[0],
                "palabra": argumentos[1],
                "tiempo": int(argumentos[2]),
                "fallas": int(argumentos[3])
            })
            
        # Ordendas el ranking por el valor del tiempo.
        rankingOrdenado = sorted(ranking, key=lambda e: e['tiempo'])
        
        # Creamos una lista para después juntarla con un join 
        rankingLista = []
        for linea in rankingOrdenado:
            rankingLista.append(
                str(len(rankingLista) + 1) + ") " +
                linea["nombre"] + " > " +
                str(linea["tiempo"]) + " segundos > " +
                "Palabra: " + linea["palabra"] + " > " +
                "Fallas: " + str(linea["fallas"])
            )
        
        return("\n".join(rankingLista))
```

```python
# Manda un mensaje por medio del sintetizador de voz
# Toma como argumento el texto a mandar
def usarSintetizador(texto):
    if not SINTETIZADOR:
        return
    
    print("\nEspera a que termine de hablar para continuar...\n")
    
    import pyttsx3
    engine = pyttsx3.init()
    engine.say(texto)
    engine.runAndWait()
```

```python
# Juego normal
# Dentro de esta función sucede todo el juego normal
# Toma como argumento la palabra que se va a usar en ese juego
def juegoNormal(palabraSeleccionada):
    if MOSTRAR_PALABRA:
        print("[DEBUG] Empezando juego normal con la palabra: " + palabraSeleccionada)
    
    tiempoInicio = math.floor(time.time())
    fallas = 0
    juegoGanado = False # Esta variable se vuelve true si se gana el juego
    letrasAdivinadas = [] # Lista para guardar las letras adivinadas
    palabraSeleccionadaEstandarizada = quitarAcentos(palabraSeleccionada) # La palabra seleccionada sin acentos o mayúsculas
    
    print(
        "\n¡Todo listo! Empezando juego normal... Mucha suerte\n" +
        "La palabra o frase tiene " + str(len(palabraSeleccionada)) + " caracteres en total."
    )
    
    # Seguir el juego mientras todavía falten fallas para completar el hombre de palo.
    while fallas < (len(HOMBRE_PALO) - 1):
        palabraEscondida = generarPalabraEscondida(palabraSeleccionada, letrasAdivinadas)
        if (palabraEscondida["palabraAdivinada"]):
            juegoGanado = True
            break
            
        print(agregarBarras(
            HOMBRE_PALO[fallas] +
            "\n" +
            palabraEscondida["palabraEscondida"]
        ))
        
        letraAdivinada = input("Escribe una letra a adivinar o la palabra completa: ")
        if len(letraAdivinada) == 0:
            print("\n¡Tienes que insertar al menos una letra o toda la palabra completa!")
            continue
            
        letraAdivinadaEstandarizada = quitarAcentos(letraAdivinada)
            
        # Si el texto es mayor a un caracter, manejarlo como si estuviera adivinando toda la palabra/frase
        if len(letraAdivinada) != 1:
            if (letraAdivinadaEstandarizada == palabraSeleccionadaEstandarizada):
                juegoGanado = True
                break
            fallas += 1
            print(
                "\n¡No! Esa no es la palabra correcta" +
                "\nAhora tienes " + str(fallas) + " falla" + ("" if fallas == 1 else "s") + " de " + str((len(HOMBRE_PALO) - 1)) + "."
            )
            continue
            
        if letraAdivinadaEstandarizada not in CARACTERES_PERMITIDOS:
            print("\n¡Ese caracter no es válido y no forma parte de la palabra! Intenta con otro...")
            continue
        if letraAdivinadaEstandarizada in letrasAdivinadas:
            print("\n¡Ya adivinaste esta letra! Intenta con otra...")
            continue
            
        letrasAdivinadas.append(letraAdivinadaEstandarizada)
        
        if letraAdivinadaEstandarizada not in palabraSeleccionadaEstandarizada:
            fallas += 1
            print(
                "\n¡Oh, no! Esa letra no aparece en la palabra. +1 falla" +
                "\nAhora tienes " + str(fallas) + " falla" + ("" if fallas == 1 else "s") + " de " + str((len(HOMBRE_PALO) - 1)) + "."
            )
            continue
            
        print("\n¡Correcto! Esa letra aparece " + str(palabraSeleccionadaEstandarizada.count(letraAdivinadaEstandarizada)) + " veces en la palabra.")
            
    # Fuera del while
    tiempoFinal = math.floor(time.time()) - tiempoInicio
    print(agregarBarras(
        "¡Fin del juego!\n\n" +
        HOMBRE_PALO[fallas]
    ))
    if juegoGanado:
        print("¡Felicidades, adivinaste la palabra!")
        usarSintetizador("¡Felicidades, adivinaste la palabra! ¿Si te dijeron que no hay premios, verdad?")
    else:
        print("¡Oh, no! Perdiste y se ahorcó el hombre de palo. :c")
        usarSintetizador(
            "¡Oh, no! Perdiste y se ahorcó el hombre de palo. carita triste. La palabra o frase correcta era " +
            palabraSeleccionada +
            ". Suerte a la próxima, jajajaja"
        )
    
    print(
        "\nResumen del juego:\n" +
        "Palabra: " + palabraSeleccionada + "\n" +
        "Fallos: " + str(fallas) + "\n" +
        "Tiempo: " + str(tiempoFinal) + " segundos"
    )
    
    # Guardar el tiempo
    if juegoGanado:
        guardarTiempo(palabraSeleccionada, tiempoFinal, fallas)
    
    respuesta = input("Pulsa ENTER para regresar al menú principal")
    return mandarMenuYEmpezar()
    
```

```python
# Juego multijugador
# Dentro de esta función sucede todo el juego multijugador
# Toma como argumento la palabra que se va a usar en ese juego y la lista de los jugadores.
def juegoMultijugador(palabraSeleccionada, nombresJugadores):
    # Creamos una lista para guardar toda la información de cada jugador.
    infoJugadores = []
    for jugador in nombresJugadores:
        infoJugadores.append({
            "nombre": jugador,
            "fallas": 0,
            "tiempo": 0,
            "letrasAdivinadas": [],
            "palabraAdivinada": False
        })
    if MOSTRAR_PALABRA:
        print("[DEBUG] Empezando juego multijugador con la palabra: " + palabraSeleccionada)

    palabraSeleccionadaEstandarizada = quitarAcentos(palabraSeleccionada) # La palabra seleccionada sin acentos o mayúsculas
    
    print(
        "\n¡Todo listo! Empezando juego multijugador... Mucha suerte\n" +
        "La palabra o frase tiene " + str(len(palabraSeleccionada)) + " caracteres en total."
    )
    
    # Seguir el juego mientras no haya algún jugador que ya tenga la palabra adivinada
    while obtenerEstadoMultijugador(infoJugadores):
        # El turno para cada jugador
        for jugador in infoJugadores:
            # Saltar el turno si el jugador ya perdió
            if jugador["fallas"] >= (len(HOMBRE_PALO) - 1):
                continue
                
            print(agregarBarras("Turno de " + jugador["nombre"] + ". Recuerda que SÓLO " + jugador["nombre"] + " puede ver esta pantalla."))
            input("Da ENTER para empezar tu turno.")
            
            # Empezar a contar el tiempo
            tiempoInicio = math.floor(time.time())
            
            # No acabar el turno del jugador hasta que tengamos una respuesta válida
            while True:
                palabraEscondida = generarPalabraEscondida(palabraSeleccionada, jugador["letrasAdivinadas"])
                
                print(agregarBarras(
                    HOMBRE_PALO[jugador["fallas"]] +
                    "\n" +
                    palabraEscondida["palabraEscondida"]
                ))

                letraAdivinada = input("Escribe una letra a adivinar o la palabra completa: ")
                if len(letraAdivinada) == 0:
                    print("\n¡Tienes que insertar al menos una letra o toda la palabra completa!")
                    continue
                    
                letraAdivinadaEstandarizada = quitarAcentos(letraAdivinada)
                
                # Si el texto es mayor a un caracter, manejarlo como si estuviera adivinando toda la palabra/frase
                if len(letraAdivinada) != 1:
                    if (letraAdivinadaEstandarizada == palabraSeleccionadaEstandarizada):
                        print("\n¡Felicidades adivinaste la palabra! Terminaremos todos los turnos de esta ronda antes de mostrar tu victoria.")
                        jugador["palabraAdivinada"] = True
                    else:
                        print(
                            "\n¡No! Esa no es la palabra correcta" +
                            "\nAhora tienes " + str(jugador["fallas"]) + " falla" + ("" if jugador["fallas"] == 1 else "s") + " de " + str((len(HOMBRE_PALO) - 1)) + "."
                        )
                        jugador["fallas"] += 1
                    break

                if letraAdivinadaEstandarizada not in CARACTERES_PERMITIDOS:
                    print("\n¡Ese caracter no es válido y no forma parte de la palabra! Intenta con otro...")
                    continue
                if letraAdivinadaEstandarizada in jugador["letrasAdivinadas"]:
                    print("\n¡Ya adivinaste esta letra! Intenta con otra...")
                    continue

                jugador["letrasAdivinadas"].append(letraAdivinadaEstandarizada)

                if letraAdivinadaEstandarizada not in palabraSeleccionadaEstandarizada:
                    jugador["fallas"] += 1
                    print(
                        "\n¡Oh, no! Esa letra no aparece en la palabra. +1 falla" +
                        "\nAhora tienes " + str(jugador["fallas"]) + " falla" + ("" if jugador["fallas"] == 1 else "s") + " de " + str((len(HOMBRE_PALO) - 1)) + "."
                    )
                    break
                print("\n¡Correcto! Esa letra aparece " + str(palabraSeleccionadaEstandarizada.count(letraAdivinadaEstandarizada)) + " veces en la palabra.")
                
                # Volver a generar la palabra escondida para ver si adivinó la palabra en este turno
                palabraEscondida = generarPalabraEscondida(palabraSeleccionada, jugador["letrasAdivinadas"])
                if (palabraEscondida["palabraAdivinada"]):
                    print("\n¡Felicidades adivinaste la palabra! Terminaremos todos los turnos de esta ronda antes de mostrar tu victoria.")
                    jugador["palabraAdivinada"] = True
                
                break
            
            # Fuera del while
            # Parar tiempo y sumarlo
            tiempoFinal = math.floor(time.time()) - tiempoInicio
            jugador["tiempo"] += tiempoFinal
            
            # Mostrar resumen y acabar turno
            print(
                "\nAsi quedó el estado de tu juego:\n" +
                HOMBRE_PALO[jugador["fallas"]]
            )
            
            if jugador["palabraAdivinada"]:
                print(
                    palabraSeleccionada +
                    "\n\nTiempo final: " + str(jugador["tiempo"]) + " segundos"
                )
            else:
                print(
                    palabraEscondida["palabraEscondida"] +
                    "\n\nTiempo por ahora: " + str(jugador["tiempo"]) + " segundos"
                )

            if jugador["fallas"] >= (len(HOMBRE_PALO) - 1):
                print(
                    "\n¡Oh, no! Perdiste y se ahorcó el hombre de palo. :c\n" +
                    "A partir de este momento, saltaremos tu turno."
                )

            respuesta = input("\n¡Fin de tu turno! Da ENTER para finalizar tu turno y borrar el texto de la pantalla.")
            borrarTexto()
        # Afuera del for (turno del jugador)
    # Afuera del while
    
    # Imprimir los ganadores
    print(generarGanadores(infoJugadores, palabraSeleccionada))
    
    respuesta = input("Pulsa ENTER para regresar al menú principal")
    return mandarMenuYEmpezar()
```
